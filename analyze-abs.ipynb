```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
gdf = gpd.read_file('MB_2021_AUST_SHP_GDA2020/MB_2021_AUST_GDA2020.shp')
# 8m
```

```python
# Check if there is a 1-1 mapping relationship between 'LGA_CODE_2023' and 'LGA_NAME_2023'
dfi_LGA = df[['LGA_CODE_2023', 'LGA_NAME_2023']].drop_duplicates().reset_index(drop=True)
dfi_SAL = df[['SAL_CODE_2021', 'SAL_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_POA = df[['POA_CODE_2021', 'POA_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_SA2 = df[['SA2_CODE_2021', 'SA2_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_SA3 = df[['SA3_CODE_2021', 'SA3_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_SA4 = df[['SA4_CODE_2021', 'SA4_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_GCCSA = df[['GCCSA_CODE_2021', 'GCCSA_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_STATE = df[['STATE_CODE_2021', 'STATE_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_AUS = df[['AUS_CODE_2021', 'AUS_NAME_2021']].drop_duplicates().reset_index(drop=True)
dfi_CHANGE = df[['CHANGE_FLAG_2021', 'CHANGE_LABEL_2021']].drop_duplicates().reset_index(drop=True)
```

```python
gdf_suburbs = gdfi[['SAL_CODE_2021', 'POA_CODE_2021', 'geometry']].dissolve(by=['SAL_CODE_2021', 'POA_CODE_2021']).reset_index()
# 3m 30s
```

```python
df_suburbs = df[['SAL_CODE_2021', 'POA_CODE_2021', 'STATE_CODE_2021', 'AUS_CODE_2021']].drop_duplicates().reset_index(drop=True)
```

```python
gdf_suburbs_area = df.groupby(['SAL_CODE_2021', 'POA_CODE_2021', 'STATE_CODE_2021', 'AUS_CODE_2021'])['AREA_ALBERS_SQKM'].sum().reset_index()
```

```python
gdf_suburbs : gpd.GeoDataFrame = gdf_suburbs.merge(gdf_suburbs_area, on=['SAL_CODE_2021', 'POA_CODE_2021'], how='left')
```

```python
gdf_suburbs = gdf_suburbs.merge(dfi_SAL, on='SAL_CODE_2021', how='left').merge(dfi_POA, on='POA_CODE_2021', how='left').merge(dfi_STATE, on='STATE_CODE_2021', how='left').merge(dfi_AUS, on='AUS_CODE_2021', how='left')
```

```python
gdf_suburbs.rename(columns={'AREA_ALBERS_SQKM': 'area_sqkm', 'SAL_NAME_2021': 'suburb', 'POA_NAME_2021': 'postcode', 'STATE_NAME_2021': 'state', 'AUS_NAME_2021': 'country'}, inplace=True)
```

```python
gdf_suburbs = gdf_suburbs[['suburb', 'postcode', 'state', 'country', 'area_sqkm', 'geometry']]
```

```python
gdf_suburbs.to_file('suburbs.geojson', driver='GeoJSON')
# 2m 30s
```
